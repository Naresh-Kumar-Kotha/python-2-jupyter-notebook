```python
import logging
from logging.handlers import RotatingFileHandler
logging.basicConfig(
        handlers=[RotatingFileHandler('./my_log.log', maxBytes=5000, backupCount=10)],
        # handlers=[RotatingFileHandler('./my_log.log')],
        level=logging.DEBUG,
        format="[%(asctime)s] %(levelname)s [%(name)s.%(funcName)s:%(lineno)d] %(message)s",
        datefmt='%Y-%m-%dT%H:%M:%S')
logger = logging.getLogger('__name__')

for i in range(1000):
    print(f"The number: {i} is {'EVEN' if i % 2 == 0 else 'ODD'}")
    logger.debug(f"The number: {i} is {'EVEN' if i % 2 == 0 else 'ODD'}")
    logger.info(f"The number: {i} is {'EVEN' if i % 2 == 0 else 'ODD'}")
    logger.warning(f"The number: {i} is {'EVEN' if i % 2 == 0 else 'ODD'}")
    logger.error(f"The number: {i} is {'EVEN' if i % 2 == 0 else 'ODD'}")
    logger.critical(f"The number: {i} is {'EVEN' if i % 2 == 0 else 'ODD'}")
```
